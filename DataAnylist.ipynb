Calculate the Average of the column

```python
avg_norm_loss = df["normalized-losses"].astype("float").mean()
print("Average of normalized losses:",avg_norm_loss)
```

Replace Nan by mean value

```python
df["normalized-losses"].replace(np.nan,avg_norm_loss,inplace=True)
```

Replace NaN in "stroke" column by mean.

```python
avg_stroke = df["stroke"].astype("float").mean(axis = 0)
print("Average of stroke:",avg_stroke)
df["stroke"].replace(np.nan,avg_stroke,inplace = True)
```

Replace NaN in "horspower" column by mean

```python
avg_horsepower = df["horsepower"].astype("float").mean(axis = 0)
print("Average of horspower:",avg_horsepower)
df["horsepower"].replace(np.nan,avg_horsepower, inplace = True)
```

Replace NaN in "bore" column by mean

```python
avg_bore = df["bore"].astype("float").mean(axis = 0)
print("Average of bore:",avg_bore)
df["bore"].replace(np.nan,avg_bore,inplace = True)
```

Replace NaN in "peak-rpm" column by mean.

```python
avg_peakrpm = df["peak-rpm"].astype("float").mean(axis = 0)
print("Average peak rpm:",avg_peakrpm)
df["peak-rpm"].replace(np.nan,avg_peakrpm,inplace = True)
```

```python
df["num-of-doors"].value_counts()
```

Replace "four" for null values

```python
df["num-of-doors"].replace(np.nan,"four",inplace=True)
```

Drop whole row with NaN in "price" column

```python
df.dropna(subset=["price"],axis=0,inplace=True)
df.reset_index(drop=True, inplace=True)
```

**Data Standardization**
Data is usually collected from different agencies with different formats.(Data standardization is also a term for a partcular type of data normalization, where we subtract the mean and divide by the stanndard deviation)

L/100km = 235 / mpg


```python
df["highway-L/100km"] = (235/df["highway-mpg"])
df.head()
```

```python
df["city-L/100km"] = 235/df["city-mpg"]
df.head()
```

Data Normalization
 
 why normalizatio?

 Normalization is the process of transforming values of several into a similar range. Typical normalizations include scaling the variable so the variable average is 0, scaling the varialble so the variance is 1,or scaling variable so  the variables values range from 0 to 1.

```python
df["length"] = df["length"]/df["length"].max()
df["width"] = df["width"]/df["width"].max()
df["height"] = df["height"]/df["height"].max()
```

```python
df[["length","width","height"]].head()
```
